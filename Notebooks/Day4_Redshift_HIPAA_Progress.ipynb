```python
# üè• Day 4: Redshift Basics + HIPAA Understanding + Project Simulation

# === 1. Setup ===
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

Load Cleaned CMS Data

```python
file_path = r'C:\Users\shazi\Healthcare_Project\output\state_readmission_summary.csv'
df = pd.read_csv(file_path)
```

Quick Preview

```python
df.head()
```

Simulate Redshift Query (Group by state)

```python
state_summary = df.groupby('state')['count_of_readm_measures_worse'].sum().reset_index()
```

Preview Result

```python
print(state_summary.head())
```

Visualization (Simulated Redshift Output)

```python
plt.figure(figsize=(14, 7))
sns.barplot(data=state_summary.sort_values('count_of_readm_measures_worse', ascending=False),
            x='state', y='count_of_readm_measures_worse', palette='coolwarm')

plt.title('Simulated Redshift: Readmission Measures by State', fontsize=16)
plt.xticks(rotation=45)
plt.ylabel('Count of Worse Readmission Measures')
plt.xlabel('State')
plt.tight_layout()
plt.savefig(r'C:\Users\shazi\Healthcare_Project\output\day4_redshift_simulation.png', dpi=300, bbox_inches='tight')
plt.show()
```
