```python
# Day 7: Athena Practice + Project Review 

# === 1. Setup ===
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

Load Cleaned Data

```python
file_path = r'C:\Users\shazi\Healthcare_Project\output\day5_hospital_readmission.csv'
df = pd.read_csv(file_path)

# Quick Preview
df.head()
```

```python
# === 3. Athena Practice Simulation (SQL style query example) ===
print("""
Simulated Athena Query:
SELECT facility_name, SUM(count_of_readm_measures_worse) AS total_readmission
FROM hospital_readmission_summary
GROUP BY facility_name
ORDER BY total_readmission DESC
LIMIT 10;
""")
```

```python
# === 4. Review Visual Outputs ===
plt.figure(figsize=(12, 7))
sns.barplot(data=df.head(10),
            x='count_of_readm_measures_worse', 
            y='facility_name', 
            palette='viridis')

plt.title('Top 10 Hospitals with Highest Readmission Measures', fontsize=16)
plt.xlabel('Count of Worse Readmission Measures', fontsize=12)
plt.ylabel('Hospital Name', fontsize=12)
plt.tight_layout()
plt.savefig(r'C:\Users\shazi\Healthcare_Project\output\day7_top10_hospitals.png', dpi=300, bbox_inches='tight')
plt.show()
```
